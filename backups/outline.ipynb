```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from typing import Optional
```

```python
img= cv2.imread("source/1/1 N NO 4_ch00.tif", 1)
#img = cv2.imread("images/Osteosarcoma_01.tif", 1)
if img is not None:
    plt.imshow(img)
```

```python
if img is not None:
    blue_channel = img[:, :, 0]
    plt.imshow(blue_channel, cmap='gray')
```

```python
#plt.hist(blue_channel.flat, bins=100, range=(0, 150))
background = blue_channel <= 50

plt.imshow(background, cmap='gray')
```

```python
nuiclei = blue_channel > 60
plt.imshow(nuiclei, cmap='gray')
```

```python
ret1, thresh1 = cv2.threshold(blue_channel, 50, 255, cv2.THRESH_BINARY)
plt.imshow(thresh1, cmap='gray')
```

```python
ret2, thresh2 = cv2.threshold(blue_channel, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
plt.imshow(thresh2, cmap='gray')
```
