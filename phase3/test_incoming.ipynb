```python
from pathlib import Path
from register_dataset import register_dataset

source_root = Path.cwd() / "source"

# Choose the dataset folder you want to inspect
dataset_folder_name = "19"          # e.g. "10", "11", "Control group", etc.
ds_folder = source_root / dataset_folder_name

print("Folder exists:", ds_folder.exists())
print("Folder:", ds_folder)

if ds_folder.is_dir():
    # Register pairs for this dataset folder
    records = register_dataset(ds_folder)
    print(f"\nDataset '{dataset_folder_name}' – pairs found:", len(records))

    for r in records:
        print("\nPair:", r.pair_name)
        print("  BF:", r.bf_path.name)
        print("  FLUO:", r.fluo_path.name)

        # List relevant metadata files for this pair
        print("  meta_core :", r.meta_core.name if r.meta_core else None)
        print("  meta_props:", r.meta_props.name if r.meta_props else None)

        print("  pixel_size_um:", r.pixel_size_um)
        print("  field_x_um   :", r.field_length_um_x)
        print("  objective    :", r.objective_name)
        print("  NA           :", r.numerical_aperture)
        print("  exposure BF/FLUO (s):", r.exposure_bf_s, r.exposure_fluo_s)

    # Optionally, list all metadata files present in the MetaData folder
    meta_dir = ds_folder / "MetaData"
    if meta_dir.is_dir():
        print("\nAll metadata files in", meta_dir)
        for p in sorted(meta_dir.glob("*.xml")):
            print(" ", p.name)
    else:
        print("\nNo MetaData folder found at", meta_dir)
```
