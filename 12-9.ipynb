```python
import cv2
import numpy as np

grey_image = r'source\12\12 N NO 1_ch00.tif'

img16 = cv2.imread(grey_image, cv2.IMREAD_UNCHANGED)
if img16 is None:
    raise ValueError(f"Image not found or unable to read image file: {grey_image}")

# Ensure grayscale
if img16.ndim == 3:
    img16 = cv2.cvtColor(img16, cv2.COLOR_BGR2GRAY)

# Normalize to [0,1] (you can keep this even though the file is uint8)
imgf = img16.astype(np.float32) / 255.0   # since your TIFF is already 8‑bit

# Blur
blur = cv2.GaussianBlur(imgf, (5, 5), 0)

# Back to 8‑bit
img8 = cv2.convertScaleAbs(blur, alpha=255.0)

# *** Inversion because objects are darker ***
img8_inv = cv2.bitwise_not(img8)

# Otsu threshold — foreground = bright objects (formerly dark)
_, thresh = cv2.threshold(
    img8_inv, 0, 255,
    cv2.THRESH_BINARY + cv2.THRESH_OTSU
)

# Find contours on binary mask
contours, hierarchy = cv2.findContours(
    thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE
)

print("Contours found:", len(contours))

# Visualize
vis = cv2.cvtColor(img8, cv2.COLOR_GRAY2BGR)  # original contrast for display
cv2.drawContours(vis, contours, -1, (0, 0, 255), 1)
cv2.imwrite("contours_overlay_inverted.png", vis)
```
