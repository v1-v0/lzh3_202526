```python
import cv2
import numpy as np
from typing import cast
```

```python
img = cv2.imread('source/1/1 N NO 5_ch00.tif', cv2.IMREAD_GRAYSCALE)

# Direct approach without cast - UMat will handle the conversion
umat = cv2.UMat(img)
sob = cv2.Sobel(umat, cv2.CV_64F, 1, 0, ksize=3)
sob_np = sob.get()  # back to numpy ndarray if needed

soblex = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3)
sobley = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=3)

gradient_magnitude = cv2.magnitude(soblex, sobley)
gradient_magnitude = cv2.convertScaleAbs(gradient_magnitude)

# Display results without blocking the kernel
print(f"Image shape: {img.shape}")
print(f"Gradient magnitude shape: {gradient_magnitude.shape}")
print(f"Gradient magnitude min/max: {gradient_magnitude.min()}/{gradient_magnitude.max()}")

# Optional: Save the result instead of displaying
# cv2.imwrite('gradient_output.png', gradient_magnitude)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.subplot(1, 2, 2)
plt.imshow(gradient_magnitude, cmap='gray')
plt.title('Gradient Magnitude')
plt.show()
```
