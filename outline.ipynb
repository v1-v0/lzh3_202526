```python
import cv2
import numpy as np
from typing import Optional
```

```python
# Read the image in grayscale
img: Optional[np.ndarray] = cv2.imread('source/1/1 N NO 5_ch00.tif', cv2.IMREAD_GRAYSCALE)

# Check if the image was loaded successfully
if img is None:
    raise FileNotFoundError("Could not load image at 'source/1/1 N NO 5_ch00.tif'")

# Option 1: UMat-based Sobel (GPU-accelerated)
umat = cv2.UMat(img)  # Convert NumPy array to UMat
soblex_umat = cv2.Sobel(umat, cv2.CV_64F, 1, 0, ksize=3)
sobley_umat = cv2.Sobel(umat, cv2.CV_64F, 0, 1, ksize=3)

# Compute gradient magnitude using UMat
gradient_magnitude_umat = cv2.magnitude(soblex_umat, sobley_umat)
gradient_magnitude = cv2.UMat.get(gradient_magnitude_umat)  # Convert back to NumPy
gradient_magnitude = cv2.convertScaleAbs(gradient_magnitude)

# Option 2: NumPy-based Sobel (CPU-based, as in your original code)
# soblex = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3)  # type: ignore
# sobley = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=3)  # type: ignore
# gradient_magnitude = cv2.magnitude(soblex, sobley)
# gradient_magnitude = cv2.convertScaleAbs(gradient_magnitude)

# Display the result
cv2.imshow('Sobel Edge Detection', gradient_magnitude)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
